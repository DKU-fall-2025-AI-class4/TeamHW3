# P2. Regression 모델의 구현과 활용

1. Linear regresstion 알고리즘의 설계에서 활용되는 least square method를 유도하시오

최소제곱법(Least Square Method)은 예측함수가 파라미터에 대해 선형일 때만 닫힌 해를 구할 수 있다.

따라서 예측함수는 다음과 같이 파라미터에 대해 선형인 함수라고 가정한다.
 $$f(x) = w_0 + w_1x$$
n개의 데이터에 대해 독립변수 $x$에 대한 행렬 $X$, 파라미터 벡터 $\theta$, 종속변수에 대한 벡터 $y$는 다음과 같이 정의된다.
$$
X = \begin{bmatrix}
1 & x_1\\
1 & x_1\\
\vdots & \vdots \\
1 & x_n\\
\end{bmatrix}
\text{ }
\theta = \begin{bmatrix}
w_0 \\ w_1
\end{bmatrix} 
\text{ }
y = \begin{bmatrix}
y_1\\
y_1\\
\vdots \\
y_n\\
\end{bmatrix}
$$

이 때 예측값에 대한 벡터 $\hat{y}$는 다음과 같이 나타난다.
$$
\hat{y} = X\theta = \begin{bmatrix}
1 & x_1\\
1 & x_1\\
\vdots & \vdots \\
1 & x_n\\
\end{bmatrix}
\begin{bmatrix}
w_0 \\ w_1
\end{bmatrix} = 

\begin{bmatrix}
w_0 + w_1x_1\\
w_0 + w_1x_2\\
\vdots\\
w_0 + w_1x_n\\
\end{bmatrix} 
$$

$\hat{y}$에 대한 평균 제곱 오차를 목적 함수로 두면 $J(\theta)$는 다음과 같은 행렬식으로 나타난다.
$$
J(\theta) = \frac{1}{n}(X\theta-y)^{\top}(X\theta-y)
$$
이를 풀어서 쓰면
$$
J(w_0,w_1) = \frac{1}{n}\sum^n_{i=1}(w_0+w_1x_i-y_i)^2
$$로 나타난다.

행렬식 $J(\theta)$를 $\theta$에 대해 미분하면
$$
{\nabla}_{\theta}J(\theta) = \frac{1}{n} · 2X^{\top}(X\theta - y) = \frac{2}{m}X^{\top}(X\theta - y)
$$

$J(\theta)$가 최소값을 가질 때, ${\nabla}_{\theta}J(\theta)$의 값이 0이 되어야 하므로
$$
X^{\top}(X\theta - y) = 0, \\
X^{\top}X\theta = X^{\top}y
$$
$X^{\top}X$가 가역일 때, $\theta$에 대해 정리하면
$$
\theta = (X^{\top}X)^{-1}X^{\top}y
$$

이것이 Least Square Method를 이용한 선형회귀의 닫힌 해이다.
